```python
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df['Severity']= label_encoder.fit_transform(df['Severity']) 
  
df['Severity'].unique() 
```

```python
def outliers(base_data_encode):
    import numpy as np
    import statistics as sts
    
    for i in base_data_encode.describe().columns:
        x=np.array(base_data_encode[i])
        p=[]
        Q1=base_data_encode[i].quantile(0.25)
        Q3=base_data_encode[i].quantile(0.75)
        IQR=Q3-Q1
        
        LTV=Q1-(1.5*IQR)
        UTV=Q3+ (1.5*IQR)
        
        for j in x:
#             print(j)
            if j<= LTV or j>= UTV:
                p.append(sts.median(x))
            else:
                p.append(j)
                
        base_data_encode[i]=p
        
    return base_data_encode
```
